```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
image = cv2.imread('/content/signal-2021-10-07-235104_005.jpeg',cv2.IMREAD_COLOR)

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

block_size = 2
ksize = 3
k = 0.04

harris_response = cv2.cornerHarris(gray_image, blockSize = block_size, ksize=ksize,k=k)

threshold = 0.01 * harris_response.max()

corner_mask = harris_response > threshold

image_with_corner = image.copy()
image_with_corner[corner_mask] = [0,0,255]

cv2_imshow(image_with_corner)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
